```python




# ================================
# Task 1: Foundational Setup & EDA
# ================================

# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

print("Libraries imported successfully.")

# Step 2: Upload dataset in Colab
from google.colab import files
uploaded = files.upload()   # Select task_1ganesh.xlsx

# Step 3: Load dataset
df = pd.read_excel("task_1ganesh.xlsx")
print("\n--- RAW DATA HEAD ---")
print(df.head())

# Step 4: Inspect dataset
print("\n--- INFO ---")
print(df.info())
print("\n--- Missing Values ---")
print(df.isnull().sum())
print("\n--- Duplicate Records ---")
print(df.duplicated().sum())

# ================================
# Data Cleaning
# ================================

# Handle missing values
df['Customer_Name'] = df['Customer_Name'].fillna("Unknown")
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce').fillna(0)
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce').fillna(0)

# Fix dates
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')

# Remove duplicates
df = df.drop_duplicates()

# Correct spelling mistakes
df['Category'] = df['Category'].replace({
    "Furnture": "Furniture",
    "Tech": "Technology"
})

# Handle negative values
df['Sales'] = df['Sales'].apply(lambda x: abs(x))
df['Profit'] = df['Profit'].apply(lambda x: abs(x))

# Final check
print("\n--- CLEANED INFO ---")
print(df.info())
print("\n--- CLEANED HEAD ---")
print(df.head())

# Export cleaned dataset
df.to_excel("task_1ganesh_cleaned.xlsx", index=False)
files.download("task_1ganesh_cleaned.xlsx")
print("\nCleaned dataset exported as task_1ganesh_cleaned.xlsx")

# ================================
# Exploratory Data Analysis (EDA)
# ================================

# Descriptive statistics
print("\n--- DESCRIPTIVE STATS ---")
print(df.describe())

# Histogram (Univariate Analysis)
plt.figure(figsize=(6,4))
sns.histplot(df['Sales'], bins=10, kde=True)
plt.title("Sales Distribution")
plt.show()

# Scatterplot (Bivariate Analysis)
plt.figure(figsize=(6,4))
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title("Sales vs Profit")
plt.show()

# Correlation Heatmap (numeric only)
numeric_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(6,4))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Numeric Columns Only)")
plt.show()

# Interactive Plotly chart
fig = px.bar(df, x='Category', y='Sales', color='Region', title="Sales by Category & Region")
fig.show()

# ================================
# Extra: Categorical Analysis
# ================================

# Count of orders by Category
plt.figure(figsize=(6,4))
sns.countplot(x='Category', data=df)
plt.title("Order Count by Category")
plt.show()

# Sales by Region
plt.figure(figsize=(6,4))
sns.barplot(x='Region', y='Sales', data=df, estimator=sum)
plt.title("Total Sales by Region")
plt.show()
```
